# Plotting in Julia

## Plotting Functions 

Adding the plotting package. 

```julia
using Pkg 
Pkg.add("Plots")      
```

```julia
using logging
Logging.disable_logging(Logging.Info)
```

```julia
using Plots
```

### sine plot

```julia
plot(
    sin,                           
    -4π, 4π,                       
    )  
```

```julia
plot(
    cos,
    -4π, 4π,
    )
```

```julia
plot(
    x -> cos(x)^5 + sin(x)^4,
    -2π, 2π
    )  
```

```julia
plot(
    x -> exp(x),
    0, 10, 
    label="-log(x)", 
    xlabel="val", 
    ylabel="-log(x)", 
    title="Behavior of Negative Log"
    )  
# Plots are by-default SVG 
```

```julia
plot(
    x -> log(x),
    0, 1000, 
    label="-log(x)", 
    xlabel="val", 
    ylabel="-log(x)", 
    title="Behavior of Negative Log"
    )  
# Plots are by-default SVG 
```

```julia
plot(
    x -> -log(x),
    0, 1000, 
    label="-log(x)", 
    xlabel="val", 
    ylabel="-log(x)", 
    title="Behavior of Negative Log"
    )  
# Plots are by-default SVG 
```

## Line plot 

```julia
x = [1:20]
```

```julia
y1, y2 = rand(20), rand(20);
```

```julia
plot(x, y1, label="y1")  
plot!(x, y2, label="y2")      
```

```julia
# Alternate way  

plot(x,   [y1, y2],    label=["First" "Second"])     
```

```julia
plot(x,   [y1, y2],  seriestype=:scatter)     
```

```julia
plot(x,   [y1, y2],  seriestype=:scatter, thickness_scaling = 0.7, opacity=0.8)     
```

```julia
y1, y2, y3, y4 = rand(20), rand(20), rand(20), rand(20)

plot(x,   [y1, y2, y3, y4], opacity=0.8, palette = :Dark2_5)

```

## Combining Multiple Plots as Subplots

```julia
x=1:10;
y = rand(10, 4)                
plot(x, y, layout = (4, 1))    
```

```julia
p1 = plot(x,y)
p2 = scatter(x,y)
p3 = plot(x,y, lw=3)
p4 = histogram(x,y)
plot(p1,p2,p3,p4, layout = (2,2))
```
